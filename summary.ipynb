```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
from matplotlib import cm
from matplotlib.colors import ListedColormap
```

```python
list_models = ["Naif","Moyenne","GAM_Gamma","GAMGamma_VAR","GAMLASSOAR","GAMPLSAR","GAMTreeAR","GAMRandomForestAR","GAMGradientBoostingAR"]
```

```python

cv_scores=pd.DataFrame()

for name in list_models:
    curr_scores = pd.read_csv("../Data/"+ name +"/Test_scores.csv")
    curr_scores = pd.DataFrame(curr_scores.mean()).transpose()
    cv_scores=pd.concat([cv_scores,curr_scores],ignore_index=True) 
    

cv_scores.insert(0,"Name",list_models)
cv_scores.to_csv("../Data/comparaison_scores.csv",index = False)
cv_scores
```
