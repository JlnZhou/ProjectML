On prend le problème de prédiction du débit moyen journalier à une semaine, le critère que l'on va minimiser est un mean square error sur toutes les stations. (discuter)

On choisis de faire de la validation croisée pour faire de la sélection de modèle, et on garde une ensemble de test.  
On a des données de 2010 à 2020 (11 années).  

On garde 2020 comme ensemble de test, et on fait 10 ensembles de validation croisée, et on utilise le split time series de sklearn (prend le début et tester sur l'année suivante), ca fait une erreur calculée sur 9 splits.

Pour les évaluations, la première prédiction est celle du 15 Janvier, que l'on prédit à partir des données du 1 au 7 Janvier, on se place le 8 à 00h00. (on laisse une semaine de données pour calculer les features, et on prédit à J+7).

```python
import pandas as pd
import numpy as np
```

```python
stations_debit = pd.read_csv("../Data/Base/Stations_Debit.csv")
mesures = pd.read_csv("../Data/Base/Mesures.csv")
mesures["Date"] = pd.to_datetime(mesures["Date"], format = "%Y/%m/%d %H:%M:%S")
```

## Ensemble de test

```python
mesures_2020 = mesures[mesures["Date"].apply(lambda x: x.year == 2020)]
```

```python
mesures_2020_X = mesures_2020.copy()
mesures_2020_X.to_csv("../Data/Base/Mesures_Test_X.csv",
                     index = False)
mesures_2020_X
```

```python
mesures_2020_Y = mesures_2020[mesures_2020["Date"].apply(lambda x: x.month == 1)][["Date"] + list(stations_debit["Code station"])].groupby(mesures_2020["Date"].apply(lambda x: x.day)).mean()
mesures_2020_Y["Jour"] = list(mesures_2020_Y.index)
mesures_2020_Y["Mois"] = 1
mesures_2020_Y = mesures_2020_Y[mesures_2020_Y["Jour"] >= 15]
for i in range(2, 13):
    mesures_mois = mesures_2020[mesures_2020["Date"].apply(lambda x: x.month == i)][["Date"] + list(stations_debit["Code station"])].groupby(mesures_2020["Date"].apply(lambda x: x.day)).mean()
    mesures_mois["Jour"] = list(mesures_mois.index)
    mesures_mois["Mois"] = i
    mesures_2020_Y = pd.concat([mesures_2020_Y, mesures_mois], axis=0)
mesures_2020_Y["Annee"] = 2020
```

```python
mesures_2020_Y.to_csv("../Data/Base/Mesures_Test_Y.csv",
                     index = False)
mesures_2020_Y
```

## Ensemble d'entrainement

```python
mesures_train = mesures[mesures["Date"].apply(lambda x: x.year < 2020)]
mesures_train
```

```python
mesures_train_X = mesures_train.copy()
mesures_train_X.to_csv("../Data/Base/Mesures_Train_X.csv",
                     index = False)
mesures_train_X
```

```python
mesures_train_Y = mesures_train[mesures_train["Date"].apply(lambda x: (x.month == 1) and (x.year == 2010))][["Date"] + list(stations_debit["Code station"])].groupby(mesures_train["Date"].apply(lambda x: x.day)).mean()
mesures_train_Y["Jour"] = list(mesures_train_Y.index)
mesures_train_Y["Mois"] = 1
mesures_train_Y["Annee"] = 2010
mesures_train_Y = mesures_train_Y[mesures_train_Y["Jour"] >= 15]
for i in range(2, 13):
    mesures_mois = mesures_train[mesures_train["Date"].apply(lambda x: (x.month == i) and (x.year == 2010))][["Date"] + list(stations_debit["Code station"])].groupby(mesures_train["Date"].apply(lambda x: x.day)).mean()
    mesures_mois["Jour"] = list(mesures_mois.index)
    mesures_mois["Mois"] = i
    mesures_mois["Annee"] = 2010
    mesures_train_Y = pd.concat([mesures_train_Y, mesures_mois], axis=0)

for annee in range(2011, 2020):
    for i in range(1, 13):
        mesures_mois = mesures_train[mesures_train["Date"].apply(lambda x: (x.month == i) and (x.year == annee))][["Date"] + list(stations_debit["Code station"])].groupby(mesures_train["Date"].apply(lambda x: x.day)).mean()
        mesures_mois["Jour"] = list(mesures_mois.index)
        mesures_mois["Mois"] = i
        mesures_mois["Annee"] = annee
        mesures_train_Y = pd.concat([mesures_train_Y, mesures_mois], axis=0)

mesures_train_Y
```

```python
mesures_train_Y.to_csv("../Data/Base/Mesures_Train_Y.csv",
                     index = False)
```

## Ensembles de validation croisée

```python
mesures_train_X = pd.read_csv("../Data/Base/Mesures_Train_X.csv")
mesures_train_X["Date"] = pd.to_datetime(mesures_train_X["Date"], format = "%Y/%m/%d %H:%M:%S")
mesures_train_X
```

```python
train_split_X = pd.DataFrame()
for i in range(9):
    train_split_X["Train_" + str(i)] = mesures_train_X["Date"].apply(lambda x: x.year <= 2010+i)
for i in range(9):
    train_split_X["Test_" + str(i)] = mesures_train_X["Date"].apply(lambda x: x.year == 2010+i+1)
train_split_X
```

```python
train_split_X.to_csv("../Data/Base/Index_CV_X.csv",
                    index=False)
```

```python
mesures_train_Y = pd.read_csv("../Data/Base/Mesures_Train_Y.csv")
mesures_train_Y
```

```python
train_split_Y = pd.DataFrame()
for i in range(9):
    train_split_Y["Train_" + str(i)] = mesures_train_Y["Annee"] <= 2010+i
for i in range(9):
    train_split_Y["Test_" + str(i)] = mesures_train_Y["Annee"] == 2010+i+1
    train_split_Y["Test_" + str(i)].iloc[(mesures_train_Y["Mois"] == 1) & (mesures_train_Y["Jour"] < 15)] = False # On enlève les 14 premiers jours
train_split_Y
```

```python
train_split_X.to_csv("../Data/Base/Index_CV_Y.csv",
                    index=False)
```
