```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
from matplotlib import cm
from matplotlib.colors import ListedColormap

from sklearn.preprocessing import SplineTransformer
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import GammaRegressor

from sklearn.decomposition import PCA
```

```python
def fct_scores_mse(Y_true, Y_pred, liste_stations):
    
    Y_true = Y_true.sort_values(by = ["Date"])
    Y_pred = Y_pred.sort_values(by = ["Date"])
    
    liste_mse = []
    for code in liste_stations:
        liste_mse.append(np.mean(np.square(Y_true[code].values - Y_pred[code].values)))
    resultat = pd.DataFrame({"Code station": liste_stations, "MSE": liste_mse})
    return resultat
```

On a 245x7 mesures météo en t, avec 7 jours d'historiques.  
Pour réduire cette dimension, on essaie de faire une PCA, puis modélisers avec une série temporelle?

```python
stations_debit = pd.read_csv("../Data/Base/Stations_Debit.csv")
liste_stations_debit = list(stations_debit["Code station"])
stations_meteo = pd.read_csv("../Data/Base/Stations_Meteo.csv")
liste_stations_meteo = list(stations_meteo["ID"].apply(lambda x: str(x).rjust(5, "0")))

mesures_train_X = pd.read_csv("../Data/Base_journalier/Mesures_Train_X.csv")
mesures_train_X["Date"] = pd.to_datetime(mesures_train_X["Date"], format = "%Y/%m/%d %H:%M:%S")
mesures_train_Y = pd.read_csv("../Data/Base_journalier/Mesures_Train_Y.csv")
mesures_train_Y["Date"] = pd.to_datetime(mesures_train_Y["Date"], format = "%Y/%m/%d %H:%M:%S")

train_split_X = pd.read_csv("../Data/Base_journalier/Index_CV_X.csv")
train_split_Y = pd.read_csv("../Data/Base_journalier/Index_CV_Y.csv")

mesures_test_X = pd.read_csv("../Data/Base_journalier/Mesures_Test_X.csv")
mesures_test_X["Date"] = pd.to_datetime(mesures_test_X["Date"], format = "%Y/%m/%d %H:%M:%S")
mesures_test_Y = pd.read_csv("../Data/Base_journalier/Mesures_Test_Y.csv")
mesures_test_Y["Date"] = pd.to_datetime(mesures_test_Y["Date"], format = "%Y/%m/%d %H:%M:%S")
```

```python
colonnes_meteo = ["Pression", "Vent_Nord", "Vent_Est", "Vitesse_vent", "Temperature", "Humidite", "Precipitations"]
tempo = []
for code in liste_stations_meteo:
    tempo += [tp + "_" + code for tp in colonnes_meteo]
colonnes_meteo_stations = tempo
colonnes_meteo_stations.sort()
mesures_train_X_meteo = mesures_train_X[["Date"] + colonnes_meteo_stations]
colonnes_meteo.sort()
colonnes_meteo
```

# PCA

```python
fig, ax = plt.subplots(figsize = (10, 10))
im = ax.imshow(mesures_train_X_meteo[colonnes_meteo_stations].corr())
fig.colorbar(im)
```

On fait de la validation croisée pour la reconstruction

```python
for curr_col in colonnes_meteo:
    print(curr_col)
    colonnes = [curr_col+"_"+code for code in liste_stations_meteo]
    for curr_compo in tqdm(range(1, len(liste_stations_meteo)+1)):
        cv_scores = pd.DataFrame({"Code station": liste_stations_meteo})
        
        for curr_split in range(9):
        
            # Entraintement
            curr_train_X = mesures_train_X[train_split_X["Train_" + str(curr_split)]].filter(regex=curr_col+".*")
            
            model_pca = PCA(n_components=curr_compo)
            model_pca.fit(curr_train_X)
            
            # Prédictions
            curr_test_X = mesures_train_X[train_split_X["Test_" + str(curr_split)]].filter(regex=curr_col+".*")
            predictions_Y = model_pca.inverse_transform(model_pca.transform(curr_test_X))
            predictions_Y = pd.DataFrame(predictions_Y, columns=colonnes)
            
            # Score
            curr_test_X["Date"] = mesures_train_X[train_split_X["Test_" + str(curr_split)]]["Date"]
            predictions_Y["Date"] = mesures_train_X[train_split_X["Test_" + str(curr_split)]]["Date"]
            curr_scores = fct_scores_mse(curr_test_X, predictions_Y, colonnes)
            cv_scores["Split_" + str(curr_split)] = curr_scores["MSE"]
        
        cv_moyen = []
        for code in liste_stations_meteo:
            score = np.mean(cv_scores[cv_scores["Code station"] == code][["Split_" + str(i) for i in range(9)]].iloc[0,:])
            cv_moyen.append(score)    
        cv_scores["Moyenne"] = cv_moyen
        cv_scores.to_csv("../Data/Modelisation_Meteo/cv_" + curr_col + "_compo_" + str(curr_compo).rjust(2, "0") + ".csv",
                     index=False)  
```

```python
for curr_col in colonnes_meteo:
    cv_moyen = pd.DataFrame()
    
    for curr_compo in range(1, len(liste_stations_meteo)+1):
        curr_moyen = {}
        cv_scores = pd.read_csv("../Data/Modelisation_Meteo/cv_" + curr_col + "_compo_" + str(curr_compo).rjust(2, "0") + ".csv")
        cv_scores["Code station"] = cv_scores["Code station"].apply(lambda x: str(x).rjust(5, "0"))
        for code in liste_stations_meteo:
            curr_moyen[code] = list(cv_scores[cv_scores["Code station"] == code]["Moyenne"])
        curr_moyen = pd.DataFrame(curr_moyen)
        curr_moyen["N_compo"] = curr_compo
        cv_moyen = pd.concat([cv_moyen, curr_moyen])
    cv_moyen.to_csv("../Data/Modelisation_Meteo/cv_moyen_" + curr_col + ".csv",
                index=False)
```
